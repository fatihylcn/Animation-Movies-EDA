```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/52000-animation-movie-details-dataset-2024/Animation_Movies.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df[df.duplicated()]
```

```python
df.drop("id", axis=1, inplace=True)
```

```python
top_ten = df[["title", "vote_count"]].sort_values("vote_count", ascending=False).head(10)

plt.figure(figsize=(15,5))
sns.barplot(data=top_ten, y="title", x="vote_count", palette="YlGnBu_r")
plt.title("Top Ten Most Voted Animated Movies")
plt.show()
```

```python
df["release_date"] = pd.to_datetime(df["release_date"], format='%Y-%m-%d')
df.loc[:,"release_year"] = df["release_date"].dt.year
df.loc[:,"release_month"] = df["release_date"].dt.month
df.loc[:,"release_day"] = df["release_date"].dt.day
```

```python
release = df["release_year"].value_counts().sort_values(ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x=release.index, y=release.values, palette="YlGnBu")
plt.title("Top Ten Years With The Most Animated Movies Released")
plt.show()
```

```python
release = df["release_month"].value_counts().sort_values(ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(x=release.index, y=release.values, palette="YlGnBu")
plt.title("Movie releases By Month")
plt.show()
```

```python
popular = df.sort_values("popularity", ascending=False)[["title", "popularity"]].head(10)
plt.figure(figsize=(15,5))
sns.barplot(popular, x="title", y="popularity", palette="YlGnBu_r")
plt.title("Top Ten Most Popular Animated Movies")
plt.xticks(rotation=45)
plt.show()
```

```python
popular = df.sort_values("revenue", ascending=False)[["title", "revenue"]].head(10)
plt.figure(figsize=(15,5))
sns.lineplot(popular, x="title", y="revenue")
plt.title("Top Ten Most Popular Animated Movies")
plt.xticks(rotation=45)
plt.show()
```

```python
frequency = {}
for row in df["spoken_languages"].str.split(","):
    for word in row:
        word = word.strip()
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
frequency_df = pd.DataFrame.from_dict([frequency]).T.sort_values(by=0, ascending=False).head(10)


plt.figure(figsize=(15,5))
sns.barplot(data = frequency_df, x=0, y=frequency_df.index, palette=sns.color_palette("YlGnBu_r"))
plt.title("Most Spoken Languages In Animated Movies")
plt.show()
```
